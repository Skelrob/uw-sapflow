```python
#plot summary 1
pdf_summary_1 =  pd.read_csv('1_summary.csv')

pdf_summary_1['unix_timestamp'] = pdf_summary_1['unix_timestamp'].map(lambda x: pd.to_datetime(x, unit='s'))

#convert UTC into Pacific Time
pdf_summary_1['unix_timestamp'] = pdf_summary_1['unix_timestamp'] - pd.Timedelta('07:00:00')

#just plot selective columns
pdf_summary_1.head()
```

```python
#just plot selective columns
# cols = ['meanhr']
# pdf_summary_1[cols].plot()

pdf_summary_1.plot(x='unix_timestamp', y='meanhr', style=".", figsize=(20,10), title='Node 1 Mean Heat Ratios')
```
